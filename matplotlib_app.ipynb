```python
import pandas as pd
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.cbook as cbook
import numpy as np
import datetime as DT
import matplotlib.dates as mdates
from datetime import date
from matplotlib.pyplot import figure

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

def plot_country_cases(country_filter):
    df=pd.read_csv("https://opendata.ecdc.europa.eu/covid19/casedistribution/csv")

    d1 = df[(df['countriesAndTerritories'] == country_filter)]
    
    plt.figure(figsize=(122,31))   
    fig, ax = plt.subplots()
    
    
    ax.plot(d1.dateRep, d1.cases)
    
    ax.set_xticks(d1.dateRep[::5])
    plt.xticks(rotation=90,size=6)
    
       
    plt.title(country_filter)
    
    plt.show()

plot_country_cases("Italy")
```
